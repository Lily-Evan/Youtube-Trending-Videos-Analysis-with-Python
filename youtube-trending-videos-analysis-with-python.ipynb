```python
from collections import Counter

# Διαχωρισμός tags (είναι χωρισμένα με '|')
tags_series = df['tags'].dropna().apply(lambda x: x.split('|'))
flat_tags = [tag.strip().lower() for sublist in tags_series for tag in sublist]
tag_counts = Counter(flat_tags)

# Top 20 tags
top_tags = dict(tag_counts.most_common(20))

plt.figure(figsize=(12,6))
sb.barplot(x=list(top_tags.values()), y=list(top_tags.keys()), palette='viridis')
plt.title("Top 20 Tags σε Trending Videos")
plt.xlabel("Συχνότητα")
plt.ylabel("Tag")
plt.show()
```

```python
keywords = ['official', 'live', 'remix']
for word in keywords:
    count = df['title'].str.lower().str.contains(word).sum()
    print(f"'{word}' εμφανίζεται σε {count} τίτλους")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Βασικά features
df_ml = df[['view_count', 'comment_count', 'categoryId', 'comments_disabled', 'ratings_disabled']].copy()

# Drop NaNs (σε περίπτωση που υπάρχουν)
df_ml.dropna(inplace=True)

# Κατηγοριοποίηση σε "low", "medium", "high" με βάση τα views
df_ml['popularity'] = pd.qcut(df_ml['view_count'], q=3, labels=['low', 'medium', 'high'])

# Είσοδοι και στόχοι
X = df_ml.drop(['view_count', 'popularity'], axis=1)
y_reg = df_ml['view_count']
y_clf = df_ml['popularity']

# Scaling για ML
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_reg, test_size=0.2, random_state=42)

reg_model = LinearRegression()
reg_model.fit(X_train, y_train)
y_pred = reg_model.predict(X_test)

print("🔢 REGRESSION:")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R² score:", r2_score(y_test, y_pred))
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=3, random_state=42)
df_ml['cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10,6))
sb.scatterplot(data=df_ml, x='comment_count', y='categoryId', hue='cluster', palette='Set2')
plt.title("Clustering Βίντεο (KMeans)")
plt.xlabel("Σχόλια")
plt.ylabel("Κατηγορία")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_clf, test_size=0.2, random_state=42)

clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_train, y_train)
y_pred = clf_model.predict(X_test)

print("📊 CLASSIFICATION:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


importances = clf_model.feature_importances_
feature_names = X.columns

feat_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feat_imp_df = feat_imp_df.sort_values('Importance', ascending=False)

plt.figure(figsize=(10,6))
sb.barplot(x='Importance', y='Feature', data=feat_imp_df, palette='magma')
plt.title("📊 Feature Importance (Random Forest)")
plt.show()
```

```python
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

df['publish_hour'] = df['publishedAt'].dt.hour
df['publish_day'] = df['publishedAt'].dt.dayofweek  # Monday = 0
df['publish_month'] = df['publishedAt'].dt.month

# Προσθέτουμε στο ML set
df_ml_extended = df_ml.copy()
df_ml_extended = df_ml_extended.join(df[['publish_hour', 'publish_day', 'publish_month']])
df_ml_extended.dropna(inplace=True)

# Αναπροσαρμογή εισόδων
X_extended = df_ml_extended.drop(['view_count', 'popularity', 'cluster'], axis=1)
y_extended = df_ml_extended['popularity']

# Scaling
X_extended_scaled = scaler.fit_transform(X_extended)

# Μοντέλο Classification ξανά
X_train, X_test, y_train, y_test = train_test_split(X_extended_scaled, y_extended, test_size=0.2, random_state=42)

clf_model.fit(X_train, y_train)
y_pred = clf_model.predict(X_test)

print("📆 Με ώρα/μέρα/μήνα δημοσίευσης:")
print(classification_report(y_test, y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Ένα απλό fully-connected NN
model = Sequential()
model.add(Dense(64, input_shape=(X_extended_scaled.shape[1],), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3 κατηγορίες: low, medium, high

# Compile
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Label encoding του y (low/medium/high -> 0/1/2)
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_nn = encoder.fit_transform(y_extended)

# Train
model.fit(X_extended_scaled, y_nn, epochs=15, batch_size=32, validation_split=0.2)
```
